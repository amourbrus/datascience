画出温度变化图

```python
# 1.准备x, y坐标数据
x = range(60)
import random
y_california = [random.uniform(12, 18) for i in x]   # random.uniform 返回随机小数
# y_california

# 2. create a canvas
plt.figure(figsize=(20,9), dpi=80)
# 3, 画折线图
plt.plot(x, y_california)
# 4, 显示图像
plt.show()
```

```python
添加自定义 x，y轴刻度 plt.xticks(x, **kwargs) plt.yticks(y, **kwargs)  x,y 表示要显示的值
```

```python
# 2. create a canvas
plt.figure(figsize=(20,9), dpi=80)

# 3, 画折线图
plt.plot(x, y_california)
# ===============================在plot之后后添加设置坐标轴、标题、图例等操作
# 构造刻度标签
x_ticks_label = ['11点{}分'.format(i) for i in x]
y = range(40)
# 修改刻度显示
plt.xticks(x[::10], x_ticks_label[::5])
plt.yticks(y[::5])
# 4, 显示图像
# ===========================在show 前加：刻度、网格、坐标轴描述及标题
plt.show()
```

绘制多条线,多次plot

```python
y_beijing = [random.uniform(1,4) for i in x]
plt.plot(x,y_california, label='y_california')
plt.plot(x,y_beijing, color='r',linestyle='--',label='y_beijing')
# 显示图例 
plt.legend(loc='best')
plt.show()
```

多个坐标系显示=== plt.subplots()

注意：plt.函数名()相当于面向过程的画图方法，axes.set_方法名()相当于面向对象的画图方法。

```python
import random

# 准备x,y数据
x = range(60)
y_california = [random.uniform(15,18) for i in x]
y_finland = [random.uniform(1,3) for i in x]
# 创建画布
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=40)

# 画图
axes[0].plot(x, y_california, label="california")
axes[1].plot(x, y_finland, label='finland',color='r', linestyle='--')
axes[0].legend()
axes[1].legend()
# ====================== 设置坐标信息，在画图后
x_ticks_label = ["11点{}分".format(i) for i in x]
y_ticks = range(40)

axes[0].set_xticks(x[::5],x_ticks_label[::5])
axes[0].set_yticks(y_ticks[::5])
axes[1].set_xticks(x[::5],x_ticks_label[::5])
axes[1].set_yticks(y_ticks[::5])
# 添加网格
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[1].grid(True, linestyle='--', alpha=0.7)
# 添加坐标轴信息和标题
axes[0].set_xlabel('time')
axes[0].set_ylabel('temperature')
axes[0].set_title('california 11-12 changes of temperature')
axes[1].set_xlabel('time')
axes[1].set_ylabel('temperature')
axes[1].set_title('finland 11-12 changes of temperature')
# ====================== 设置坐标信息

plt.show()
```

### 柱状图 bar

> 分类数据对比的场景：数据统计，用户数量对比分析

```python
# 对比每部电影的票房收入
```

```python
# 准备数据
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
x = range(len(movie_name))
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]
# 创建画布
plt.figure(figsize=(20,8),dpi=60)
# 绘图
plt.bar(x,y,width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])
# 修改刻度显示
plt.xticks(x, movie_name)
plt.grid(linestyle='--',alpha=0.6)
# title
plt.title('电影片票房收入对比')
plt.show()
```
